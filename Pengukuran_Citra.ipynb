```python
# Instal pustaka yang diperlukan
!pip install pillow matplotlib

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

def image_dimensions(image_path):
    """Mengembalikan dimensi gambar (tinggi x lebar)."""
    with Image.open(image_path) as img:
        return img.size  # Mengembalikan tuple (tinggi, lebar)

def grayscale_histogram(image_path):
    """Mengembalikan histogram keabuan gambar."""
    with Image.open(image_path) as img:
        img = img.convert('L')  # Mengkonversi ke citra grayscale
        histogram, _ = np.histogram(img, bins=256, range=(0, 255))
        return histogram

# Unggah gambar ke Google Colab Anda
from google.colab import files
uploaded = files.upload()

# Dapatkan dimensi gambar
image_path = '/content/sample_image.png'  # Ubah ini sesuai dengan nama gambar Anda
dimensions = image_dimensions(image_path)
print(f"Dimensi Gambar (tinggi x lebar): {dimensions}")

# Tampilkan histogram keabuan gambar
histogram = grayscale_histogram(image_path)
plt.figure(figsize=(10, 5))
plt.title('Histogram Keabuan Gambar')
plt.xlabel('Intensitas Keabuan')
plt.ylabel('Frekuensi')
plt.plot(histogram, color='black')
plt.grid(True)
plt.show()
```
